```python
# ------------------------------
# 1. DATA EXTRACTION
# ------------------------------
state_mapping={"andaman-&-nicobar-islands":"Andaman & Nicobar Islands",
                                 "andhra-pradesh":"Andhra Pradesh",
                                 "arunachal-pradesh":"Arunachal Pradesh",
                                 "assam":"Assam",
                                 "bihar":"Bihar",
                                 "chandigarh":"Chandigarh",
                                 "chhattisgarh":"Chhattisgarh",
                                 "dadra-&-nagar-haveli-&-daman-&-diu":"Dadra and Nagar Haveli and Daman and Diu",
                                 "delhi":"Delhi",
                                 "goa":"Goa",
                                 "gujarat":"Gujarat",
                                 "haryana":"Haryana",
                                 "himachal-pradesh":"Himachal Pradesh",
                                 "jammu-&-kashmir":"Jammu & Kashmir",
                                 "jharkhand":"Jharkhand",
                                 "karnataka":"Karnataka",
                                 "kerala":"Kerala",
                                 "ladakh":"Ladakh",
                                 "lakshadweep":"Lakshadweep",
                                 "madhya-pradesh":"Madhya Pradesh",
                                 "maharashtra":"Maharashtra",
                                 "manipur":"Manipur",
                                 "meghalaya":"Meghalaya",
                                 "mizoram":"Mizoram",
                                 "nagaland":"Nagaland",
                                 "odisha":"Odisha",
                                 "puducherry":"Puducherry",
                                 "punjab":"Punjab",
                                 "rajasthan":"Rajasthan",
                                 "sikkim":"Sikkim",
                                 "tamil-nadu":"Tamil Nadu",
                                 "telangana":"Telangana",
                                 "tripura":"Tripura",
                                 "uttar-pradesh":"Uttar Pradesh",
                                 "uttarakhand":"Uttarakhand",
                                 "west-bengal":"West Bengal"
                               }

#Importing pandas,json & os functionality
import pandas as pd
import json
import os
from sqlalchemy import create_engine
import streamlit as st
import plotly.express as px

#Listing the states as in Aggregated Statewise Transaction folder
path="C:/VS_Python/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)

#Extraction of Aggregated Statewise Transaction Data into Dataframe
clm={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transaction_type'].append(Name)
              clm['Transaction_count'].append(count)
              clm['Transaction_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe Aggregated Transaction
Agg_Trans=pd.DataFrame(clm)
Agg_Trans["State"]=Agg_Trans["State"].replace(state_mapping)


#Listing the states as in Aggregated User folder
path1="C:/VS_Python/pulse/data/aggregated/user/country/india/state/"
Agg_state_list1=os.listdir(path1)


#Extraction of Aggregated Statewise User Data into Dataframe
clm1={'State':[], 'Year':[],'Quarter':[],'Registered_User':[],'App_Opens':[]}
clm1_1={    'State': [], 'Year': [], 'Quarter': [],    'Brand': [], 'User_Count': [], 'Percentage': []}

for i in Agg_state_list1:
    p_i=path1+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            aggregated = D['data']['aggregated']
            reg_user = aggregated['registeredUsers']
            app_opens = aggregated['appOpens']

            clm1['State'].append(i)
            clm1['Year'].append(j)
            clm1['Quarter'].append(int(k.strip('.json')))
            clm1['Registered_User'].append(reg_user)
            clm1['App_Opens'].append(app_opens)

            users_by_device = D['data'].get('usersByDevice') 
            if users_by_device:   # only loop if not None / not empty
                for device in users_by_device:
                    clm1_1['State'].append(i)
                    clm1_1['Year'].append(j)
                    clm1_1['Quarter'].append(int(k.strip('.json')))
                    clm1_1['Brand'].append(device['brand'])
                    clm1_1['User_Count'].append(device['count'])
                    clm1_1['Percentage'].append(device['percentage'])

#Succesfully created a dataframe Aggregated User
Agg_User=pd.DataFrame(clm1)
Agg_User_Device=pd.DataFrame(clm1_1)
Agg_User["State"]=Agg_User["State"].replace(state_mapping)
Agg_User_Device["State"]=Agg_User_Device["State"].replace(state_mapping)


#Listing the states as in Aggregated Insurance folder
path2="C:/VS_Python/pulse/data/aggregated/insurance/country/india/state/"
Agg_state_list2=os.listdir(path2)


#Extraction of Aggregated Statewise Insurance Data into Dataframe
clm2={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list2:
    p_i=path2+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm2['Transaction_type'].append(Name)
              clm2['Transaction_count'].append(count)
              clm2['Transaction_amount'].append(amount)
              clm2['State'].append(i)
              clm2['Year'].append(j)
              clm2['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe Aggregated Insurance
Agg_Insurance=pd.DataFrame(clm2)
Agg_Insurance["State"]=Agg_Insurance["State"].replace(state_mapping)


#Listing the states as in Map Transaction folder
path3="C:/VS_Python/pulse/data/map/transaction/hover/country/india/state/"
Agg_state_list3=os.listdir(path3)


#Extraction of Statewise Map Transaction Data into Dataframe
clm3={'State':[], 'Year':[],'Quarter':[],'District':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list3:
    p_i=path3+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['hoverDataList']:
              Name=z['name']
              count=z['metric'][0]['count']
              amount=z['metric'][0]['amount']
              clm3['District'].append(Name)
              clm3['Transaction_count'].append(count)
              clm3['Transaction_amount'].append(amount)
              clm3['State'].append(i)
              clm3['Year'].append(j)
              clm3['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe Map Transaction
Map_Transaction=pd.DataFrame(clm3)
Map_Transaction["State"]=Map_Transaction["State"].replace(state_mapping)



#Listing the states as in Map User folder
path4="C:/VS_Python/pulse/data/map/user/hover/country/india/state/"
Agg_state_list4=os.listdir(path4)


#Extraction of Map User Data into Dataframe
clm4={'State':[], 'Year':[],'Quarter':[],'District':[],'Registered_User':[],'App_Opens':[],}

for i in Agg_state_list4:
    p_i=path4+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['hoverData'].items():
              name=z[0]
              reg_user=z[1]['registeredUsers']
              app_open=z[1]['appOpens']
              clm4['District'].append(name)
              clm4['Registered_User'].append(reg_user)
              clm4['App_Opens'].append(app_open)
              clm4['State'].append(i)
              clm4['Year'].append(j)
              clm4['Quarter'].append(int(k.strip('.json')))

#Succesfully created a dataframe Map User
Map_User=pd.DataFrame(clm4)
Map_User["State"]=Map_User["State"].replace(state_mapping)



#Listing the states as in Map Insurance folder
path5="C:/VS_Python/pulse/data/map/insurance/hover/country/india/state/"
Agg_state_list5=os.listdir(path5)


#Extraction of Map Insurance Data into Dataframe
clm5={'State':[], 'Year':[],'Quarter':[],'District':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list5:
    p_i=path5+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['hoverDataList']:
              Name=z['name']
              count=z['metric'][0]['count']
              amount=z['metric'][0]['amount']
              clm5['District'].append(Name)
              clm5['Transaction_count'].append(count)
              clm5['Transaction_amount'].append(amount)
              clm5['State'].append(i)
              clm5['Year'].append(j)
              clm5['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe Map Insurance
Map_Insurance=pd.DataFrame(clm5)
Map_Insurance["State"]=Map_Insurance["State"].replace(state_mapping)



#Listing the states as in Top Transaction folder
path6="C:/VS_Python/pulse/data/top/transaction/country/india/state/"
Agg_state_list6=os.listdir(path6)


#Extraction of Top Transaction - Pincode Data into Dataframe
clm6={'State':[], 'Year':[],'Quarter':[],'Pincode':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list6:
    p_i=path6+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['pincodes']:
              Name=z['entityName']
              amount=z['metric']['amount']
              count=z['metric']['count']
              clm6['Pincode'].append(Name)
              clm6['Transaction_count'].append(count)
              clm6['Transaction_amount'].append(amount)
              clm6['State'].append(i)
              clm6['Year'].append(j)
              clm6['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe Top Transaction
Top_Transaction=pd.DataFrame(clm6)
Top_Transaction["State"]=Top_Transaction["State"].replace(state_mapping)


#Listing the states as in Top User folder
path7="C:/VS_Python/pulse/data/top/user/country/india/state/"
Agg_state_list7=os.listdir(path7)


#Extraction of Top User - Pincode Data into Dataframe
clm7={'State':[], 'Year':[],'Quarter':[],'Pincode':[], 'Registered_User':[]}

for i in Agg_state_list7:
    p_i=path7+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['pincodes']:
              Name=z['name']
              reg_users=z['registeredUsers']
              clm7['Pincode'].append(Name)
              clm7['Registered_User'].append(reg_users)
              clm7['State'].append(i)
              clm7['Year'].append(j)
              clm7['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe Top User
Top_User=pd.DataFrame(clm7)
Top_User["State"]=Top_User["State"].replace(state_mapping)



#Listing the states as in Top Insurance folder
path8="C:/VS_Python/pulse/data/top/insurance/country/india/state/"
Agg_state_list8=os.listdir(path8)


#Extraction of Top Insurance - Pincode Data into Dataframe
clm8={'State':[], 'Year':[],'Quarter':[],'Pincode':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in Agg_state_list8:
    p_i=path8+i+"/"
    Agg_st=os.listdir(p_i)
    for j in Agg_st:
        p_j=p_i+j+"/"
        Agg_yr=os.listdir(p_j)
        for k in Agg_yr:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['pincodes']:
              Name=z['entityName']
              count=z['metric']['count']
              amount=z['metric']['amount']
              clm8['Pincode'].append(Name)
              clm8['Transaction_count'].append(count)
              clm8['Transaction_amount'].append(amount)
              clm8['State'].append(i)
              clm8['Year'].append(j)
              clm8['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe Top Insurance
Top_Insurance=pd.DataFrame(clm8)
Top_Insurance["State"]=Top_Insurance["State"].replace(state_mapping)

# ------------------------------
# 2. DATABASE CREATION IN SQL
# ------------------------------

engine = create_engine('mysql+mysqlconnector://root:Welcome123@127.0.0.1:3306/phonepe')

Agg_Trans.to_sql('aggregate_transaction',con=engine,if_exists='replace',index=False)
Agg_User.to_sql('aggregate_user',con=engine,if_exists='replace',index=False)
Agg_User_Device.to_sql('aggregate_user_device',con=engine,if_exists='replace',index=False)
Agg_Insurance.to_sql('aggregate_insurance',con=engine,if_exists='replace',index=False)
Map_Transaction.to_sql('map_transaction',con=engine,if_exists='replace',index=False)
Map_User.to_sql('map_user',con=engine,if_exists='replace',index=False)
Map_Insurance.to_sql('map_insurance',con=engine,if_exists='replace',index=False)
Top_Transaction.to_sql('top_transaction',con=engine,if_exists='replace',index=False)
Top_User.to_sql('top_user',con=engine,if_exists='replace',index=False)
Top_Insurance.to_sql('top_insurance',con=engine,if_exists='replace',index=False)
```
